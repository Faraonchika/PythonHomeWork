# Data description

```python
import random
```

```python
class warrior():
    health = 0
    strength = 0
    name = ''
    
    def __init__(self, newName):
        self.name = newName
        self.health = random.randint(100, 150)
        self.strength = random.randint(20, 30)
    
    def get_health(self):
        return self.health
    
    def get_strength(self):
        return self.strength
    
    def get_damage(self, hit_power):
        self.health -= hit_power
        if self.health >= 0:
            print(self.name, "получил", hit_power, "урона. Осталось", self.health, "здоровья.")
        else:
            print(self.name, "получил", hit_power, "урона и погиб.")
```

```python
class warrior_with_sheild(warrior):
    protection = random.randint(5, 10)
    
    def get_damage(self, hit_power):
        damage = hit_power - self.protection
        self.health -= damage
        if self.health >= 0:
            print(self.name, "получил", damage, "урона. Осталось", self.health, "здоровья.")
        else:
            print(self.name, "получил", damage, "урона и погиб.")
```

```python
class warrior_expert(warrior):
    def get_strength(self):
        r = random.random()
        if r <= 0.2:
            new_power = 2 * self.strength
            return new_power
        else:
            return self.strength
```

# Task 1

```python
warrior1 = warrior('TheChosenOne')
warrior2 = warrior_with_sheild('TheChosenTwo')
warrior3 = warrior_expert('TheChosenThree')
```

```python
print(warrior1.get_strength())
print(warrior2.get_strength())
s = warrior3.get_strength()
if s > 30:
    s = s / 2
print(s)
```

# Task 3

```python
army1 = [warrior('Воин' + str(i + 1) + ' из армии 1') for i in range(4)] + \
[warrior_with_sheild('Воин со щитом' + str(i + 1) + ' из армии 1') for i in range(4)] + \
[warrior_expert('Воин эксперт' + str(i + 1) + ' из армии 1') for i in range(2)]
```

```python
army2 = [warrior('Воин' + str(i + 1) + ' из армии 2') for i in range(4)] + \
[warrior_with_sheild('Воин со щитом' + str(i + 1) + ' из армии 2') for i in range(4)] + \
[warrior_expert('Воин эксперт' + str(i + 1) + ' из армии 2') for i in range(2)]
```

```python
print("Исходные характеристики армий:")
print("Суммарное здоровье - армия 1:", sum([now.get_health() for now in army1]), "армия 2:",sum([now.get_health() for now in army2]))
print("Средний урон - армия 1:", sum([now.get_strength() for now in army1]) / 10, "армия 2:", sum([now.get_strength() for now in army2]) / 10)
```

```python
def make_fight(warrior1, warrior2):
    turn = 0
    while warrior1.get_health() >= 0 and warrior2.get_health() >= 0:
        if turn == 0:
            warrior2.get_damage(warrior1.get_strength())
            turn = 1
        elif turn == 1:
            warrior1.get_damage(warrior2.get_strength())
            turn = 0
```

```python
# Let's make war!
# С функцией выглядит красивее, но не выполняется условие очередности...
while len(army1) > 0 and len(army2) > 0:
    r1 = random.randint(0, len(army1) - 1)
    r2 = random.randint(0, len(army2) - 1)
    make_fight(army1[r1], army2[r2])
    if army1[r1].get_health() < 0:
        army1.pop(r1)
    else:
        army2.pop(r2)
if len(army1) == 0:
    print('Победила армия 2')
else:
    print('Победила армия 1')
```

```python
# Let's make war! №2
# Не так кратко, но верно
turn = 0
while len(army1) > 0 and len(army2) > 0:
    r1 = random.randint(0, len(army1) - 1)
    r2 = random.randint(0, len(army2) - 1)
    warrior1 = army1[r1]
    warrior2 = army2[r2]
    while warrior1.get_health() >= 0 and warrior2.get_health() >= 0:
        if turn == 0:
            warrior2.get_damage(warrior1.get_strength())
            turn = 1
        elif turn == 1:
            warrior1.get_damage(warrior2.get_strength())
            turn = 0
    if army1[r1].get_health() < 0:
        army1.pop(r1)
    else:
        army2.pop(r2)
if len(army1) == 0:
    print('Победила армия 2')
else:
    print('Победила армия 1')
```
